```python
# First, I will create a simple recommender that just recommends movies based on user ratings
# I will use IMDB's weighted rating system
# WR = ((v / (v + m))(R)) + ((m / (v + m))(C))
# where v = number of user votes for movie
# m = minimum number of votes to qualify for chart
# R = average rating of movie
# C = mean vote across entire userbase

C = md['vote_average'].mean()
C

```

```python
# For m, we will use the 95th percentile to be cut-off mark

m = md['vote_count'].quantile(0.95)
m
```

```python
qual_movies = md.copy().loc[md['vote_count'] >= m]
qual_movies
```

```python
# Perform the weighted rating on qual_movies

def weighted_rating(movies, m=m, C=C):
    v = movies['vote_count']
    R = movies['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

# Add a new column "scores" to qual_movies that determines how good the movie is 
qual_movies['score'] = qual_movies.apply(weighted_rating, axis=1)
```

```python
# Sort the movies in descending order based on their score
qual_movies = qual_movies.sort_values('score', ascending=False)
#Print the top 15 movies
qual_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
# Imports TfidfVectorizer, which is Term Frequency-Inverse Document Frequency, basically how often a word appears 
# and the number of documents that have the word. A word is more significant if it appears a lot on a page, and not 
# on many of the pages
from sklearn.feature_extraction.text import TfidfVectorizer

# Creates a tfidf object, and gets rid of words like 'the', 'and'
tfidf = TfidfVectorizer(stop_words='english')

# Replaces NaN values with ''
md['overview'] = md['overview'].fillna('')
# Creating the TF-IDF matrix by transforming and fitting it to the data
tfidf_matrix = tfidf.fit_transform(md['overview'])

tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Construct a reverse map of indices and movie titles
indices = pd.Series(md.index, index=md['title']).drop_duplicates()
indices
```

```python
cosine_sim
```

```python
cosine_sim.shape
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get index of the movie that corresponds to title
    idx = indices[title]

    # Get cosine similarity scores of all movies with the movie 
    # sim_scores = list(enumerate(cosine_sim[idx])) is efficient way
    sim_scores = []
    for i in range(len(indices)):
        sim_scores.append([i, cosine_sim[i][idx]])
        
    # Sort movies based on the similarity scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies by slicing it to get elements 1-11 as 0 would be the movie itself
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return top 10 most similar movies to movie in question
    return md['title'].iloc[movie_indices]

get_recommendations('Finding Dory')
```

```python
# Using credits, genres, and keywords to build a recommender

# Convert the strings that look like tuples into actual tuples 
from ast import literal_eval

# The features we will use to recommend movies based on 
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    md[feature] = md[feature].apply(literal_eval)
    
# Function to get the director from the cast names
def getDirector(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan # Case where there is no director, so return NaN

# Returns the list of top 3 elements or entire list if there is 3 or less elements in the list
def getList(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        # Check if more than 3 elements exist and if there are more than 3, then reduce the names to only the top 3
        if len(names) > 3:
            names = names[:3]
        return names

    # Return empty list in case of missing data
    return []

```

```python
# Define new director, cast, genres and keywords features after applying function to put them in the right form
md['director'] = md['crew'].apply(getDirector)

features = ['cast', 'keywords', 'genres']
for feature in features:
    md[feature] = md[feature].apply(getList)
    
md[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
# Function to convert all strings to lower case and get rid of space in between first and last names so that for the
# vectorization, names with the same first name but different last name are treated as different names
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        # Check if director exists. If it does not, return the empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
        
# Apply the clean_data function to all of the features
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    md[feature] = md[feature].apply(clean_data)
```

```python
# Now, using all the features amassed so far, we can create the soup that will be fed into the word vectorizer
def create_metadata_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
md['soup'] = md.apply(create_metadata_soup, axis=1)
```

```python
# Now that we have the soup, we can begin to build our recommender 

# Import CountVectorizer and create the count matrix. Not using TF-IDF because an actor appearing in multiple movies
# should not take away from the significance of them being in a certain movie
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english') # TF-IDF did not need stop_words because of the Inverse Document Frequency took care of it
count_matrix = count.fit_transform(md['soup'])

# Compute Cosine Similarity matrix based on count matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
# Collaborative Filtering 
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
reader = Reader()

ratings.head(5)
```

```python
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
algo = SVD()
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)# This means we have divided our dataset into 
# 5 groups(folds) which will either act as training sets or the sets used to test our model and 
# evaluates performance of model
```

```python
# Based on the cross_validate function, we obtained a RMSE of 0.8972, which is satisfactory for our purposes.
# We can now begin to train our model

trainset = data.build_full_trainset()
algo.fit(trainset)
```
